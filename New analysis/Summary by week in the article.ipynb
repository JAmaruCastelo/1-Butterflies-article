```python
import pandas as pd
import datetime as dt
```

```python
data=pd.read_excel("D:/Butterfly article correction/reanalisis/Material suplementario 2.xlsx")
```

```python
data.head(5)
```

```python
# change to datetime the column needed for the analysis
data["Date"]=pd.to_datetime(data["Date"],format="%Y-%m-%d")
data=data.sort_values(by="Date")# sort the values used in the analysis
```

```python
final=pd.DataFrame()
while len(data)>0:
    # choose the five minimun dates to the analysis
    e=pd.Timedelta(days=1)
    fechas=[min(data["Date"]),min(data["Date"])+pd.Timedelta(days=1), min(data["Date"])+pd.Timedelta(days=2),min(data["Date"])+pd.Timedelta(days=3),min(data["Date"])+pd.Timedelta(days=4)]
    # extract the data bases with this dates to continue with the analysis
    conca=pd.concat([data[data["Date"] == fechas[0]],data[data["Date"] == fechas[1]],data[data["Date"] == fechas[2]],
                     data[data["Date"] == fechas[3]],data[data["Date"] == fechas[4]]])
    conca["Date_C"]=str(min(fechas))+"to"+str(max(fechas))
    # create a new code to the analysis to include the variation
    conca["Code"]=conca["Date_C"]+conca["Tramp"]+"-"+conca["Height"]
    # maintain the indixes used
    indices=conca.index
    # grouped by the code and use the richness and abundance
    values=conca.groupby("Code")["Richness", "Abundance"].sum()
    # create the dictionary to see the type of forest
    diccionario={"P":"SLR","D":"CCR","S":"PCR"}
    # create a new data bases
    valores=pd.DataFrame(values)
    valores["Year"]=conca["Year"].unique()[0]
    valores["Month"]=conca["Month"].unique()[0]
    valores["Tramp"]=[e[-5:-2] for e in valores.index]
    valores["Precipitation"]=conca["Precipitation "].unique()[0] # this is called valores
    valores["Forest"]=[diccionario[e] for e in [e[-5:-4] for e in valores.index]]
    final=pd.concat([final, valores])
    data=data.drop(indices)
```

```python
final["Height"]=[e[-1:] for e in final.index]
```

```python
final.to_excel("segunda_version.xlsx")
```
